```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeRegressor
```

```python
data = pd.read_csv('/content/drive/MyDrive/Machine Learning Project/Decision Tree Regressor/Position_Salaries.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
X = data.iloc[:, 1:-1].values
Y = data.iloc[:, -1].values

print(X, Y)
```

```python
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X, Y)
```

```python
regressor.predict([[6.5]])
```

```python
plt.scatter(X, Y, color = 'red')
plt.plot(X, regressor.predict(X), color = 'blue')
plt.title('Salary Prediction by Decision Tree Regression')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

It is best fitted in for training data and same data is used for prediction. So it is meaningless.


```python
plt.scatter(X, Y, color = 'red')
X_smooth = np.arange(min(X), max(X), 0.1)
X_smooth = X_smooth.reshape(len(X_smooth), 1)
plt.plot(X_smooth, regressor.predict(X_smooth), color = 'blue')
plt.title('Salary Prediction by Decision Tree Regression')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```
