## Importing Libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
data = pd.read_csv('/content/drive/MyDrive/Machine Learning Project/Polinomial Regression/Position_Salaries.csv')
data.shape
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
X = data.iloc[:, 1:-1].values
Y = data.iloc[:, -1].values

print(X.shape, Y.shape)
```

```python
print(X)
```

## Apply Linear Regression

```python
LR1 = LinearRegression()
LR1.fit(X, Y)
```

## Analyse by the Graph

```python
plt.scatter(X, Y, color = "red")
plt.plot(X, LR1.predict(X), color = "blue")
plt.title("Linear Regression Result")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

```python
Y_pred = LR1.predict([[6.5,]])
print(Y_pred)
```

prediction single value [[]]

```python
print(X, Y)
```

```python
PF = PolynomialFeatures(degree = 2)
X_poly = PF.fit_transform(X)

LR2 = LinearRegression()
LR2.fit(X_poly, Y)
```

```python
plt.scatter(X, Y, color = "red")
plt.plot(X, LR2.predict(X_poly), color = "blue")
plt.title("Polynomial Regression (Degree 2)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

```python
PF3 = PolynomialFeatures(degree = 3)
X_poly3 = PF3.fit_transform(X)

LR3 = LinearRegression()
LR3.fit(X_poly3, Y)
```

```python
plt.scatter(X, Y, color = "red")
plt.plot(X, LR3.predict(X_poly3), color = "blue")
plt.title("Polynomial Regression (Degree 3)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

```python
PF4 = PolynomialFeatures(degree = 4)
X_poly4 = PF4.fit_transform(X)

LR4 = LinearRegression()
LR4.fit(X_poly4, Y)
```

```python
plt.scatter(X, Y, color = "red")
plt.plot(X, LR4.predict(X_poly4), color = "blue")
plt.title("Polynomial Regression (Degree 4)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

```python
LR4.predict(PF4.transform([[6.5,]]))
```

```python
print(X)
```

```python
X_smooth = np.arange(X.min(), X.max(), 0.1)
X_smooth = X_smooth.reshape((len(X_smooth), 1))
X_smooth_poly = PF4.transform(X_smooth)
plt.scatter(X, Y, color = "red")
plt.plot(X_smooth, LR4.predict(X_smooth_poly), color = "blue")
plt.title("Polynomial Regression (Degree 4) with smooth")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```
